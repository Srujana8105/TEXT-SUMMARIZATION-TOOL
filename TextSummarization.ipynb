```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from heapq import nlargest
nltk.download('punkt_tab')
def summarize_text(text, num_sentences=3):
    stop_words = set(stopwords.words("english"))

    words = word_tokenize(text.lower())


    freq = {}
    for word in words:
        if word.isalnum() and word not in stop_words:
            freq[word] = freq.get(word, 0) + 1

    sentences = sent_tokenize(text)
    sentence_scores = {}

    for sentence in sentences:
        for word in word_tokenize(sentence.lower()):
            if word in freq:
                sentence_scores[sentence] = sentence_scores.get(sentence, 0) + freq[word]

    summary_sentences = nlargest(num_sentences, sentence_scores, key=sentence_scores.get)

    return " ".join(summary_sentences)


if __name__ == "__main__":
    text = """
    Artificial Intelligence is transforming industries across the world.
    It helps automate tasks, analyze large datasets, and improve decision-making.
    AI is widely used in healthcare, finance, education, and transportation.
    However, ethical concerns and data privacy remain major challenges.
    Responsible AI development is crucial for long-term benefits.
    """

    summary = summarize_text(text, num_sentences=2)
    print("Summary:")
    print(summary)
```
